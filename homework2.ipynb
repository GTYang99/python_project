```python
# 下載檔案成為二維list

import requests
import io
import csv

url = 'http://opendataap2.e-land.gov.tw/./resource/files/2021-03-11/132147d68273721d057a4e0247d83995.csv'
response = requests.get(url)
response.encoding = 'utf-8'
# print(response.text)
with io.StringIO(response.text) as like_file:
    csv_reader = csv.reader(like_file)
    csv_list = list(csv_reader)
print(csv_list)
```

```python
# 下載的文字，儲存檔案
with open('Yilan.csv','w',newline='') as file:
    file.write(response.text)
#不能使用list存檔，只能使用純文字檔案存檔
```

```python
import tkinter as tk
import requests

class WINDOWS(tk.Tk):
    def __init__(self):
        super().__init__()

    def download(self):
        url = 'http://opendataap2.e-land.gov.tw/./resource/files/2021-03-11/132147d68273721d057a4e0247d83995.csv'
        response = requests.get(url)
        response.encoding = 'utf-8'
        print(response)

def main():
    window = WINDOWS()
    window.mainloop()

if __name__ == '__main__':
    main()
```
